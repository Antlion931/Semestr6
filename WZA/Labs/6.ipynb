```python
def order(X, Y):
    if len(X) != len(Y):
        return False
    
    for (x, y) in zip(X, Y):
        if x > y:
            return False
    
    return True
```

```python
order([1, 2, 3], [2, 2, 2])
```

```python
def min_in_A(A):
    result = []

    for x in A:
        early_break_in_loop = False
        for r in result:
            if order(r, x):
                early_break_in_loop = True
                break
            if order(x, r):
                result.remove(r)

        if not early_break_in_loop:
            result += [x]

    return result

            
```

```python
more_or_eq_11_but_less_then_100 = []

for x in range(1, 100):
    for y in range(1, 100):
        if x*y >= 11 and x*y < 100:
            more_or_eq_11_but_less_then_100 += [[x, y]]

more_or_eq_11_but_less_then_100
```

```python
min_in_A(more_or_eq_11_but_less_then_100)
```

```python
circle_on_10x10_with_radius_5 = []

for x in range(5, 16):
    for y in range(5, 16):
        if (x - 10)**2 + (y - 10)**2 <= 25:
            circle_on_10x10_with_radius_5 += [[x, y]]

circle_on_10x10_with_radius_5
```

```python
min_in_A(circle_on_10x10_with_radius_5)
```
